```python
import pandas as pd
import sqlite3

file_path = r"C:\Users\æ®µ\Desktop\724\Renewable_Energy_Adoption.csv"
df = pd.read_csv(file_path)
df.index = range(1, len(df) + 1)

pd.get_option('display.max_rows')
pd.set_option('display.max_rows', 300)

# Calculate the government investment component
country_averages = {}

for country in df['Country'].unique():
    country_data = df[df['Country'] == country]
    avg_investment = country_data['Government Investment (Million USD)'].mean()
    country_averages[country] = avg_investment

for index, row in df.iterrows():
    if pd.isnull(row['Government Investment (Million USD)']):
        current_country = row['Country']
        avg_value = country_averages[current_country]
        df.at[index, 'Government Investment (Million USD)'] = avg_value


def none_values(row):
    energy_cols = ['Solar', 'Wind', 'Hydro', 'Geothermal', 'Biomass', 'Other Renewables']

    if pd.isnull(row['Renewable Energy (%)']):
        total = 0
        for col in energy_cols:
            if pd.isnull(row[col]):
                total += 0
            else:
                total += row[col]
        row['Renewable Energy (%)'] = total

    else:
        missing_count = 0
        missing_col = None
        known_sum = 0
        for col in energy_cols:
            if pd.isnull(row[col]):
                missing_count += 1
                missing_col = col
            else:
                known_sum += row[col]

        if missing_count == 1:
            row[missing_col] = row['Renewable Energy (%)'] - known_sum
        else:

    return row

df = df.apply(none_values, axis=1)

df.head(30)

```
